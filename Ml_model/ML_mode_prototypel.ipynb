```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/Mzuhaibkhan/Hackathon_ncs_model/refs/heads/main/ayurvedic_diseases_and_treatments_cleaned.csv')

# Normalize the symptoms in the dataset to lowercase for case-insensitive matching
df['Symptoms'] = df['Symptoms'].str.lower()

# Prepare the features and target
X = df[['Disease', 'Symptoms']]
y = df['Ayurvedic Treatment']

# Encode categorical features
X_encoded = pd.get_dummies(X, columns=['Disease', 'Symptoms'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize and train the model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Evaluate the model
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy:.2f}")

# Function to predict disease and treatment from user input
def predict_treatment(user_symptoms):
    # Convert user symptoms to lowercase for case-insensitive matching
    user_symptoms = [symptom.lower() for symptom in user_symptoms]

    # Filter rows where at least one symptom matches
    potential_matches = df[df['Symptoms'].apply(lambda x: any(symptom in x for symptom in user_symptoms))]

    if not potential_matches.empty:
        print("\nPotential Matches Based on Provided Symptoms:")
        for index, row in potential_matches.iterrows():
            matched_symptoms = [symptom for symptom in user_symptoms if symptom in row['Symptoms']]
            print(f"- Disease: {row['Disease']}")
            print(f"  Ayurvedic Treatment: {row['Ayurvedic Treatment']}")
            print(f"  Matched Symptoms: {', '.join(matched_symptoms)}\n")
    else:
        # Suggest diseases when there are insufficient matches
        print("No exact matches found. Showing diseases with similar symptoms:")
        # Display diseases with the highest overlap of symptoms
        similar_matches = df[df['Symptoms'].apply(lambda x: len(set(user_symptoms).intersection(set(x.split(", ")))) > 0)]
        if not similar_matches.empty:
            for index, row in similar_matches.iterrows():
                print(f"- Disease: {row['Disease']}")
                print(f"  Ayurvedic Treatment: {row['Ayurvedic Treatment']}")
                print(f"  Related Symptoms: {row['Symptoms']}\n")
        else:
            print("No related diseases found. Please refine your input.")

# Take user input
user_symptoms = input("Enter symptoms separated by commas (e.g., Pimples on face, Oily skin, Redness): ").split(", ")
predict_treatment(user_symptoms)
```
